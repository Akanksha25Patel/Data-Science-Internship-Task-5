Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like
    
    # Conditional Logic
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Taking number of posts from user
n = int(input("Enter number of posts: "))

likes = []

# Taking likes for each post
for i in range(n):
    like = int(input(f"Enter likes for post {i+1}: "))
    likes.append(like)

# Function Call
analyze_engagement(likes)
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def check_rainfall(rainfall_list, required_level):
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain
    
    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)
    
    # Conditional Logic
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # Output
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)


# Taking input from user
n = int(input("Enter number of months: "))
rainfall_data = []

for i in range(n):
    rain = float(input(f"Enter rainfall for month {i+1}: "))
    rainfall_data.append(rain)

required = float(input("Enter required rainfall level: "))

# Function Call
check_rainfall(rainfall_data, required)
```

Problem 4: Social Media – Duplicate Account Detection


```python
def detect_duplicate_accounts(usernames):
    
    seen = set()
    duplicates = set()
    
    # Loop to find duplicates
    for name in usernames:
        if name in seen:
            duplicates.add(name)
        else:
            seen.add(name)
    
    if len(duplicates) > 0:
        print("Duplicate Accounts Found: Yes")
        print("Duplicate Usernames:", list(duplicates))
    else:
        print("Duplicate Accounts Found: No")


# Taking input from user
n = int(input("Enter number of usernames: "))
username_list = []

for i in range(n):
    name = input(f"Enter username {i+1}: ")
    username_list.append(name)

# Function Call
detect_duplicate_accounts(username_list)
```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(prices):
    
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)


# Example Data (Market prices)
crop_prices = [1500, 2500, 1800, 3200, 1900]

# Function Call
filter_premium_crops(crop_prices)
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def check_attendance(attendance_list):
    
    total_classes = len(attendance_list)
    attended_classes = 0
    
    for status in attendance_list:
        if status == 1:
            attended_classes += 1
    
    percentage = (attended_classes / total_classes) * 100
    
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    # Output
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


# Taking input from user
n = int(input("Enter total number of classes: "))
attendance = []

print("Enter attendance for each class (1 for Present, 0 for Absent):")

for i in range(n):
    status = int(input(f"Class {i+1}: "))
    attendance.append(status)

# Function Call
check_attendance(attendance)
```

Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_bill(units):
    
    # Bill Calculation using slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status


# Taking input from user
units_consumed = int(input("Enter electricity units consumed: "))

# Function Call
total_bill, usage_status = calculate_bill(units_consumed)

# Output
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
